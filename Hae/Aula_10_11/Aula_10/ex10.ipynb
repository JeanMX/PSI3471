```python
import os; os.environ['TF_CPP_MIN_LOG_LEVEL']='3'
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import optimizers
import numpy as np
import sys
```

```python
import numpy as np
def le(nomearq):
     with open(nomearq,"r") as f:
         linhas=f.readlines()
     linha0=linhas[0].split()
     nl=int(linha0[0]); nc=int(linha0[1])
     a=np.empty((nl,nc),dtype=np.float32)
     for l in range(nl):
        linha=linhas[l+1].split()
        for c in range(nc):
             a[l,c]=np.float32(linha[c])
     return a

ax = le("ax.txt"); ay = le("ay.txt")
qx = le("qx.txt"); qy = le("qy.txt")
```

```python
from keras.layers import Dropout
model = Sequential();
model.add(Dense(30, activation='relu', input_dim=len(ax[0])))
model.add(Dropout(rate=0.25))
model.add(Dense(50, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
sgd=optimizers.SGD(learning_rate=10);
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(ax, ay, epochs=500, batch_size=10, verbose=True)
```

```python
teste = model.evaluate(qx,qy)
print("Custo e acuracidade de teste:",teste)
```

```python
qp=model.predict(qx)
print("Classificacao de teste:\n",qp)
qp = qp.argmax(axis=-1)
print("Rotulo de saida:\n",qp)
```
